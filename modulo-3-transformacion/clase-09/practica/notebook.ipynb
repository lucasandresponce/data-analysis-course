### 4) Limpieza: aplicar las decisiones

Ahora sí, implementen las decisiones del Paso 3. Recuerden:

- Primero eliminen duplicados
- Después manejen nulos
- Después conviertan tipos

```python
# Limpieza de pacientes
df_pacientes_limpio = df_pacientes.copy()
# eliminar duplicado
df_pacientes_limpio = df_pacientes_limpio.drop_duplicates(subset=['nombre','dni'])
# rellenar 
df_pacientes_limpio['obra_social'] = df_pacientes_limpio['obra_social'].fillna('Sin Obra Social')
df_pacientes_limpio = df_pacientes_limpio.fillna('Sin Datos')
#Conversion
df_pacientes_limpio['fecha_nacimiento'] = pd.to_datetime(df_pacientes_limpio['fecha_nacimiento'], errors='coerce')


# Limpieza de médicos
df_medicos_limpio = df_medicos.copy()
# elimino duplicados
df_medicos_limpio= df_medicos_limpio.drop_duplicates(subset=['matricula'])

# Limpieza de turnos
df_turnos_limpio = df_turnos.copy()
# eliminar duplicado
df_turnos_limpio = df_turnos_limpio.drop_duplicates(subset=['paciente_id','medico_id','fecha','hora'])
#conversion 
df_turnos_limpio['costo'] = pd.to_numeric(df_turnos_limpio['costo'], errors='coerce')
df_turnos_limpio['fecha'] = pd.to_datetime(df_turnos_limpio['fecha'], errors='coerce')
#rellenar
df_turnos_limpio['costo'] = df_turnos_limpio['costo'].fillna(0)
df_turnos_limpio = df_turnos_limpio.fillna('Sin Datos')

```

> **Por qué `.copy()`:** cuando hacen `df_pacientes_limpio = df_pacientes`, no están creando una copia — están apuntando al mismo objeto. Cualquier cambio en `df_pacientes_limpio` modifica también `df_pacientes`. `.copy()` crea una copia independiente, así conservan los datos originales para comparar.

Al terminar, impriman un resumen de qué cambió en cada tabla:

```python
# comparacion 
print(f"Filas originales en Turnos: {len(df_turnos)}")
print(f"Filas tras limpiar duplicados: {len(df_turnos_limpio)}")
print("-" * 50)

print(f"Filas originales en Medicos: {len(df_medicos)}")
print(f"Filas tras limpiar duplicados: {len(df_medicos_limpio)}")
print("-" * 50)

print(f"Filas originales en Turnos: {len(df_pacientes)}")
print(f"Filas tras limpiar duplicados: {len(df_pacientes_limpio)}")
print("-" * 50)
```

### 5) — Análisis: responder las preguntas del director

Con los datos limpios, respondan estas cuatro preguntas. Para cada una, combinen SQL (para traer los datos necesarios) con Pandas (para calcular o presentar el resultado):

**Pregunta 1:** ¿Cuántos turnos realizó cada médico? Mostrá el resultado ordenado de mayor a menor.

**Pregunta 2:** ¿Cuál es el costo total facturado por especialidad médica? Considerá solo los turnos con estado `'realizado'`.

**Pregunta 3:** ¿Qué obra social tiene más pacientes registrados? ¿Y cuántos pacientes no tienen obra social?

**Pregunta 4:** ¿Cuántos turnos fueron cancelados o marcados como `'ausente'`? ¿Qué médico tuvo más cancelaciones?


```python
#Pregunta 1

import sqlite3

# Volvemos a conectar con la base de datos
conexion = sqlite3.connect('medicare.db')

# cosulta
query1 = """
SELECT m.nombre, COUNT(t.id) as cantidad_turnos
FROM turnos t
JOIN medicos m ON t.medico_id = m.id
GROUP BY m.nombre
ORDER BY cantidad_turnos DESC
"""

df_ranking_actividad = pd.read_sql(query1, conexion)
df_ranking_actividad
```

```python
#Pregunta 2
query2 = """
SELECT m.especialidad, SUM(t.costo) as total_facturado
FROM turnos t
JOIN medicos m ON t.medico_id = m.id
WHERE t.estado = 'realizado'
GROUP BY m.especialidad
ORDER BY total_facturado DESC
"""
df_facturacion_especialidad = pd.read_sql(query2, conexion)
df_facturacion_especialidad
```

```python
#pregunta 3

# Contamos las frecuencias en la columna limpia
conteo_obras = df_pacientes_limpio['obra_social'].value_counts()

print(f"Obra social con más pacientes: {conteo_obras.idxmax()}")
print(f"Pacientes sin obra social: {conteo_obras.get('Sin Obra Social', 0)}")
```

```python
# pregunta 4

# Total de turnos no concretados usando Pandas
total_no_concretados = len(df_turnos_limpio[df_turnos_limpio['estado'].isin(['cancelado', 'ausente'])])

# Médico con más cancelaciones usando SQL
query4 = """
SELECT m.nombre, COUNT(t.id) as total_cancelados
FROM turnos t
JOIN medicos m ON t.medico_id = m.id
WHERE t.estado = 'cancelado'
GROUP BY m.nombre
ORDER BY total_cancelados DESC
LIMIT 1
"""
medico_critico = pd.read_sql(query4, conexion)

#Cierro conexion 
conexion.close()

print(f"Total de turnos cancelados o ausentes: {total_no_concretados}")
print(f"El médico con más cancelaciones es: {medico_critico['nombre'].values}")
```
